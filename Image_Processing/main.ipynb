```python
import matplotlib.image as img
import matplotlib.pyplot as plt
import numpy as np
```

```python

PATH = 'img/855.jpg'
original = img.imread(PATH)

def toGrayScale(path):
    #grayscale formula 
    #gray = 0.2126 * red + 0.7152 * green + 0.0722 * blue
    image = img.imread(path)
    height = len(image)
    width = len(image[0])
    #creating an array for grayscaled pixels
    gray = np.empty([height, width, 3], dtype = np.uint8)
    #looping through pixels
    for i in range(height):
        for j in range(width):
            #[arr_group][arr_row][arr_col] 
            gray[i][j] = int(image[i][j][0] * 0.2126  + image[i][j][1] * 0.7152 + image[i][j][2] * 0.0722)
    return gray
   
def binarize(path):
    #grayscaled image
    image = toGrayScale(path)
    height = len(image)
    width = len(image[0])
    #calculate the initial T value with the mean of all elements
    t = int(np.mean(image))
    #splitting the pixels for greater and smaller than the threshold
    g1 = np.empty([height, width, 3], dtype = np.uint8)
    #looping through pixels
    for i in range(height):
        for j in range(width):
            for k in range(3):
                #darker than threshold
                if image[i][j][k] <= t:
                    g1[i][j][k] = 0
                else:
                    g1[i][j][k] = 255
    return g1
    
    
def show(original, grayscaled, binary):
    f, axarr = plt.subplots(1,3, figsize = (18, 13)) 
    axarr[0].axis('off')
    axarr[0].set_title("Original")
    axarr[0].imshow(original)
    
    axarr[1].axis('off')
    axarr[1].set_title('Grayscaled')
    axarr[1].imshow(grayscaled)
    
    axarr[2].axis('off')
    axarr[2].set_title('Binarized')
    axarr[2].imshow(binary)
    
```

```python
grayscale = toGrayScale(PATH)
binary = binarize(PATH)
show(original, grayscale, binary)
```
