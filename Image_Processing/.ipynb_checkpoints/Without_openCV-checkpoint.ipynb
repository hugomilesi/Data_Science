```python
import matplotlib.image as img
import matplotlib.pyplot as plt
import numpy as np
```

```python
PATH = '855.jpg'

def toGrayScale(path):
    #grayscale formula 
    #gray = 0.2126 * red + 0.7152 * green + 0.0722 * blue
    image = img.imread(path)
    height = len(image)
    width = len(image[0])
    #creating an array with all the pixels and colors
    gray = np.empty([height, width, 3], dtype = np.uint8)

    for i in range(height):
        for j in range(width):
            #[arr_group][arr_row][arr_col] 
            gray[i][j] = int(image[i][j][0] * 0.2126  + image[i][j][1] * 0.7152 + image[i][j][2] * 0.0722)
    
    return gray
   
def binarize(path):
    image = toGrayScale(path)
    height = len(image)
    width  = len(image[0])
    binary = np.empty([height, width, 3], dtype = np.uint8)
   
    for i in range(height):
        for j in range(width):
            zero = image[i][j][0] or image[i][j][1] or image[i][j][2] < 200
            if zero:
                binary[i][j][0], binary[i][j][1], binary[i][j][2] = 255, 255, 255
            else:
                binary[i][j][0], binary[i][j][1], binary[i][j][2] = 0, 0, 0
    return binary
                
```

```python
grayscaled = toGrayScale(PATH)
plt.imshow(grayscaled)

binarized = binarize(PATH)
plt.imshow(binarized)
```
